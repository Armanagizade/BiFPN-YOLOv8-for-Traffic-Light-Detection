```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv_path = 'yolov8n results/finalYOLOv8n_40_640_round7/results.csv'  
```

```python
yolov8_map50_list=[]
yolov8_box_loss_list = []
yolov8_cls_loss_list = []
yolov8_dfl_loss_list = []


df = pd.read_csv(csv_path)
for j in range (np.shape(df['metrics/mAP50(B)'])[0]):
    yolov8_map50_list.append(df['metrics/mAP50(B)'][j]*100)
    yolov8_box_loss_list.append(df['train/box_loss'][j])
    yolov8_cls_loss_list.append(df['train/cls_loss'][j])
    yolov8_dfl_loss_list.append(df['train/dfl_loss'][j])
    
```

```python
np.shape(yolov8_map50_list)
```

```python
csv_path = 'improved results/improvedYOLOv8n_200_640_round13/results.csv'  
```

```python
improvedyolov8_map50_list=[]
improvedyolov8_box_loss_list = []
improvedyolov8_cls_loss_list = []
improvedyolov8_dfl_loss_list = []


df = pd.read_csv(csv_path)
for j in range (np.shape(df['metrics/mAP50(B)'])[0]):
    improvedyolov8_map50_list.append(df['metrics/mAP50(B)'][j]*100)
    improvedyolov8_box_loss_list.append(df['train/box_loss'][j])
    improvedyolov8_cls_loss_list.append(df['train/cls_loss'][j])
    improvedyolov8_dfl_loss_list.append(df['train/dfl_loss'][j])
    
```

```python
np.shape(improvedyolov8_map50_list)
```

```python
fig, axs = plt.subplots(1,2, figsize=(20, 5))
axs[0].plot(np.arange(200),yolov8_map50_list, label='BiFPN-Yolov8n')
axs[0].plot(np.arange(200),improvedyolov8_map50_list, label='Yolov8n')

axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('mAP-50')
axs[0].scatter(199, yolov8_map50_list[-1], color='red')  # Mark the last mAP50 point
axs[0].text(199-6, yolov8_map50_list[-1]+1, f'{yolov8_map50_list[-1]}', fontsize=10, color='red')
axs[0].scatter(199, improvedyolov8_map50_list[-1], color='red')  # Mark the last mAP50 point
axs[0].text(199-6, improvedyolov8_map50_list[-1]-2.5, f'{improvedyolov8_map50_list[-1]}', fontsize=10, color='red')
axs[0].legend()

axs[1].plot(np.arange(200),yolov8_box_loss_list, label='BiFPN-Yolov8n Train Box Loss')
axs[1].plot(np.arange(200),improvedyolov8_box_loss_list, label='Yolov8n Train Box Loss')

axs[1].plot(np.arange(200),yolov8_cls_loss_list, label='BiFPN-Yolov8n Train class Loss')
axs[1].plot(np.arange(200),improvedyolov8_cls_loss_list, label='Yolov8n Train class Loss')



axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('losses')
axs[1].legend()
```
